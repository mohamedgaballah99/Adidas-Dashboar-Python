```python
methods_sales=df.groupby('Sales Method')['Total Sales'].sum().astype('int').to_frame()
methods_sales
```

```python
#Ø§Ø­Ù…Ø§Ù„Ù‰ Ø§Ù„Ù…Ø¨ÙŠØ¹Ø§Øª Ø¨Ø§Ù„Ù†Ø³Ø¨Ø© Ù„Ø·Ø±Ù‚ Ø§Ù„Ø¨ÙŠØ¹
#Ø§Ù„Ø§Ù„ÙˆØ§Ù†
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999'] 
# Ù†Ø¬Ù…Ø¹ Ø¥Ø¬Ù…Ø§Ù„ÙŠ Ø§Ù„Ù…Ø¨ÙŠØ¹Ø§Øª Ø­Ø³Ø¨ Ø·Ø±ÙŠÙ‚Ø© Ø§Ù„Ø¨ÙŠØ¹
methods_sales = df.groupby('Sales Method')['Total Sales'].sum().reset_index()

# Ù†Ø±Ø³Ù… Ø§Ù„Ù€ Pie Chart Ø¨Ø¹Ø¯ Ø§Ù„ØªØ¬Ù…ÙŠØ¹
plt.pie(methods_sales['Total Sales'], labels=methods_sales['Sales Method'], autopct='%1.1f%%',colors=colors)# Ù„Ø¹Ø±Ø¶ Ø§Ù„Ù†Ø³Ø¨ Ø§Ù„Ù…Ø¦ÙˆÙŠØ© 
plt.title('Sales Method Sales')
plt.show()
```

```python
#Ø§Ø­Ù…Ø§Ù„Ù‰ Ø§Ù„ÙˆØ­Ø¯Ø§Øª Ø§Ù„Ù…Ø¨Ø§Ø¹Ø© Ø¨Ø§Ù„Ù†Ø³Ø¨Ø© Ù„Ù„Ù…Ù†ØªØ¬
product_sales=df.groupby('Product')['Units Sold'].sum().reset_index().sort_values(by='Units Sold' ,ascending=True).head(5)
product_sales_chart=product_sales
```

```python
product_sales_chart=product_sales.plot( x='Product', y='Units Sold' ,kind='barh',figsize=(8,3),title= "Top 5 Product Units Sold"
                                 ,ylabel='',xlabel='Count of Units',color='#db6216')
# Ù†Ø¶ÙŠÙ Ø§Ù„Ù‚ÙŠÙ… Ø¹Ù„Ù‰ Ø§Ù„Ø£Ø¹Ù…Ø¯Ø©
for i, (value, label) in enumerate(zip(product_sales['Units Sold'], product_sales['Product'])):
    product_sales_chart.text(
        value + (max(product_sales['Units Sold']) * 0.01),  # Ù†Ø­Ø±Ùƒ Ø§Ù„Ø±Ù‚Ù… ÙŠÙ…ÙŠÙ† Ø§Ù„Ø¹Ù…ÙˆØ¯ Ø´ÙˆÙŠØ©
        i,                                                  # Ù…ÙˆØ¶Ø¹Ù‡ Ø¹Ù„Ù‰ Ù…Ø­ÙˆØ± Y
        f'{value:,}',                                       # Ù†Ø¹Ø±Ø¶ Ø§Ù„Ø±Ù‚Ù… Ø¨ÙÙˆØ§ØµÙ„
        va='center',
        fontsize=9
    )

plt.show()
```

```python
#Ù†Ø³Ø¨Ø© Ø§Ù„Ø±Ø¨Ø­  Ø¨Ø§Ù„Ù†Ø³Ø¨Ø© State
top_state_sales=df.groupby('State')['Operating Profit'].sum().astype('int64').reset_index().sort_values(by='Operating Profit' ,ascending=True).head(6)
top_state_sales_chart=top_state_sales
top_state_sales_chart
```

```python

top_state_sales_chart=top_state_sales.plot(x='State',y='Operating Profit',kind='barh',figsize=(8,3),color='#db6216',ylabel='')
# Ù†Ø¶ÙŠÙ Ø§Ù„Ù‚ÙŠÙ… Ø¹Ù„Ù‰ Ø§Ù„Ø£Ø¹Ù…Ø¯Ø©
for i, (value, label) in enumerate(zip(top_state_sales['Operating Profit'], top_state_sales['State'])):
    top_state_sales_chart.text(
        value + (max(top_state_sales['Operating Profit']) * 0.01),  
        i,                                                 
        f'{value:,}',                                       
        va='center',
        fontsize=9,
    )

plt.show()
```

```python
# ØªØ­ÙˆÙŠÙ„ Ø§Ù„ØªØ§Ø±ÙŠØ® Ù„ØµÙŠØºØ© datetime
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])

# ØªØ¬Ù…ÙŠØ¹ Ø§Ù„Ù…Ø¨ÙŠØ¹Ø§Øª ÙƒÙ„ Ø±Ø¨Ø¹ Ø³Ù†Ø© (Quarter End)
quarter_sales = df.resample('QE', on='Invoice Date')['Total Sales'].mean().astype('int').reset_index()

# Ù†Ø­ÙˆÙ„ Ø§Ù„ØªØ§Ø±ÙŠØ® Ù„Ø§Ø³Ù… Ø§Ù„Ø±Ø¨Ø¹ (Ø²ÙŠ Q1 2024)
quarter_sales['Quarter_Label'] = quarter_sales['Invoice Date'].dt.to_period('Q').astype(str)

# Ø±Ø³Ù… Ø§Ù„Ø®Ø·
plt.figure(figsize=(8,4))
plt.plot(quarter_sales['Quarter_Label'], quarter_sales['Total Sales'], marker='o', color='dodgerblue')
plt.title('Average Sales per Quarter')
plt.grid(True, linestyle='--', alpha=0.5)

# Ø¥Ø¶Ø§ÙØ© Ø§Ù„Ù‚ÙŠÙ… ÙÙˆÙ‚ Ø§Ù„Ù†Ù‚Ø§Ø·
for x, y in zip(quarter_sales['Quarter_Label'], quarter_sales['Total Sales']):
    plt.text(
        x, 
        y + (max(quarter_sales['Total Sales']) * 0.01),  # Ù†Ø­Ø±Ùƒ Ø§Ù„Ø±Ù‚Ù… ÙÙˆÙ‚ Ø§Ù„Ù†Ù‚Ø·Ø© Ø´ÙˆÙŠØ©
        f'{y:,}', 
        ha='center', 
        va='bottom', 
        fontsize=9
    )

plt.tight_layout()
plt.show()
```
